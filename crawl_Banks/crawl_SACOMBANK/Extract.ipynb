```python
import openpyxl
import csv
import json
import requests
import re
import pandas as pd
import time
from time import sleep
import pyautogui as pag
from bs4 import BeautifulSoup
from selenium import webdriver
import undetected_chromedriver as uc
from selenium.webdriver import Edge
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import NoSuchAttributeException
```

```python
options = uc.ChromeOptions()
options.headless = False
options.add_argument("--disable-blink-features=AutomationControlled")
driver = uc.Chrome(version_main=135)
driver.get("https://www.sacombank.com.vn/trang-chu/danh-sach-atm-dgd.html")
```

```python
page_source = BeautifulSoup(driver.page_source, 'html.parser')
location_div = page_source.find('div', 'location-content')
locations = location_div.find_all('div', 'location')
print(len(locations))
```

```python
type_list = []
name_list = []
address_list = []
phone_list = []
workinghour_list = []
latitude_list = []
longitude_list = []
```

```python
for point in locations:
    # Loại chi nhánh (từ <div class="tag">)
    tag = point.find("div", class_="tag")
    type_value = tag.text.strip() if tag else None
    
    # Tên chi nhánh
    title = point.find("div", class_="location--title")
    name = title.text.strip() if title else None
    
    # Địa chỉ
    description = point.find("div", class_="location--description")
    address = description.text.strip() if description else None
    
    # Số điện thoại (từ <a> trong <div class="phone-number">)
    phone_div = point.find("div", class_="phone-number")
    phone = phone_div.find("a").text.strip() if phone_div and phone_div.find("a") else None
    
    # Thời gian hoạt động
    working_time = point.find("div", class_="working-time")
    workinghour = working_time.text.replace("Thời gian hoạt động", "").strip() if working_time else None
    
    # Vĩ độ
    latitude_div = point.find("div", class_="d-none latitude")
    latitude = latitude_div.text.strip() if latitude_div else None
    
    # Kinh độ
    longitude_div = point.find("div", class_="d-none longtitude")
    longitude = longitude_div.text.strip() if longitude_div else None
    
    # Thêm vào danh sách
    type_list.append(type_value)
    name_list.append(name)
    address_list.append(address)
    phone_list.append(phone)
    workinghour_list.append(workinghour)
    latitude_list.append(latitude)
    longitude_list.append(longitude)

# In kết quả
print(f"Số lượng type: {len(type_list)}")
print(type_list)
print(f"Số lượng name: {len(name_list)}")
print(name_list)
print(f"Số lượng address: {len(address_list)}")
print(address_list)
print(f"Số lượng phone: {len(phone_list)}")
print(phone_list)
print(f"Số lượng workinghour: {len(workinghour_list)}")
print(workinghour_list)
print(f"Số lượng latitude: {len(latitude_list)}")
print(latitude_list)
print(f"Số lượng longitude: {len(longitude_list)}")
print(longitude_list)

# Lưu vào CSV (tùy chọn)
df = pd.DataFrame({
    "Type": type_list,
    "Name": name_list,
    "Address": address_list,
    "Phone": phone_list,
    "WorkingHour": workinghour_list,
    "Latitude": latitude_list,
    "Longitude": longitude_list
})
df.to_csv("SACOMBANK(ĐGD+CDM+ATM+STM).csv", index=False, encoding="utf-8-sig")
```
