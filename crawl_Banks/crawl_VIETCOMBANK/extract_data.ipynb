```python
with open("DATA/province.json", "r", encoding="utf-8") as file:
    json_data = json.load(file)
codes = [province["code"] for province in json_data["provinces"]]
# df = pd.DataFrame({"code": list(set(codes))})
df = pd.DataFrame({"code": list(set(codes))})
display(df)
```

```python
chip_list = [
    "{E57A2832-1BB1-4AAC-BCB6-5F2E30F723DC}",
    "{B21940D1-3408-4051-8532-1B0EA9827088}",
    "{5968EF13-5ADC-46EF-B94F-A7EA7C00977E}",
    "{C6C2234D-4450-4D37-9F4A-465AC9666A08}"
]
for chip in chip_list:
    chip_responses = []
    for province_code in df['code']:
        data = {
            "Chip": chip,
            "Latitude": "",
            "Longitude": "",
            "ProvinceCode": province_code,
            "l": "vi-VN"
        } 
        try:
            response = requests.post(url, headers=headers, json=data)
            if response.status_code == 200:
                try:
                    response_data = response.json()
                    response_data = {
                        "Chip": chip,
                        "ProvinceCode": province_code,
                        "data": response_data
                    }
                    chip_responses.append(response_data)
                    print(f"ThÃ nh cÃ´ng cho Chip={chip}, ProvinceCode={province_code}")
                except ValueError:
                    error_response = {
                        "Chip": chip,
                        "ProvinceCode": province_code,
                        "error": "Pháº£n há»i khÃ´ng pháº£i JSON há»£p lá»",
                        "response_text": response.text
                    }
                    chip_responses.append(error_response)
                    print(f"Pháº£n há»i khÃ´ng pháº£i JSON há»£p lá» cho Chip={chip}, ProvinceCode={province_code}: {response.text}")
            else:
                error_response = {
                    "Chip": chip,
                    "ProvinceCode": province_code,
                    "error": f"Lá»i {response.status_code}",
                    "response_text": response.text
                }
                chip_responses.append(error_response)
                print(f"Lá»i: {response.status_code} cho Chip={chip}, ProvinceCode={province_code}: {response.text}")
                
        except requests.RequestException as e:
            error_response = {
                "Chip": chip,
                "ProvinceCode": province_code,
                "error": f"Lá»i khi gá»­i yÃªu cáº§u: {str(e)}"
            }
            chip_responses.append(error_response)
            print(f"Lá»i khi gá»­i yÃªu cáº§u cho Chip={chip}, ProvinceCode={province_code}: {e}")
        time.sleep(1)
    try:
        file_name = f"{chip}.json"
        with open(file_name, "w", encoding="utf-8") as f:
            json.dump(chip_responses, f, ensure_ascii=False, indent=4)
        print(f"ÄÃ£ lÆ°u pháº£n há»i vÃ o {file_name}")
    except OSError as e:
        print(f"Lá»i khi lÆ°u file {file_name}: {e}")
```
