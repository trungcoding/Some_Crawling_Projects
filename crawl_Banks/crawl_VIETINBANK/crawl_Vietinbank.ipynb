```python
import openpyxl
import csv
import json
import requests
import re
import pandas as pd
import time
from time import sleep
import pyautogui as pag
from bs4 import BeautifulSoup
from selenium import webdriver
import undetected_chromedriver as uc
from selenium.webdriver import Edge
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import NoSuchAttributeException
```

```python
options = uc.ChromeOptions()
options.headless = False  # Hiển thị trình duyệt (False) để debug dễ hơn
options.add_argument("--disable-blink-features=AutomationControlled")
driver = uc.Chrome(version_main=134)
driver.get("https://www.vietinbank.vn/mang-luoi-chi-nhanh-atm")
```

```python
soup = BeautifulSoup(driver.page_source)
address = []
name = []
```

```python
shop_address = soup.find_all('dt', 'mt-1 text-left text-xs')
shop_name = soup.find_all('dt', 'text-left text-sm font-bold')
shop_address_list = [addr.text.strip() for addr in shop_address]
shop_name_list = [name.text.strip() for name in shop_name]
print(shop_address_list)
print(len(shop_address_list))
print(shop_name_list)
print(len(shop_name_list))
```

```python
latitudes = []
longitudes = []
urls = soup.find_all('a', 'focus:border-white focus:border-0 focus:ring-0 focus:outline-none')
for url in urls:
    href = url['href']
    if "maps/dir//" in href:
        try:
            coords = href.split("maps/dir//")[1]
            lat, lon = coords.split(",")
            latitudes.append(float(lat))
            longitudes.append(float(lon))
        except:
            continue
print("Latitudes:", latitudes)
print(len(latitudes))
print("Longitudes:", longitudes)
print(len(longitudes))
```

```python
df = pd.DataFrame({
    'Name': shop_name_list,
    'Address': shop_address_list,
    'Latitude': latitudes,
    'Longitude': longitudes
})

# Ghi ra file CSV
df.to_csv('output.csv', index=False, encoding='utf-8 sig')
```
