```python
import pandas as pd 
import json
import numpy as np
from unidecode import unidecode
```

```python
input = "DATA/Add_Foody_HaiDuong.csv"
output = "DATA/Add_Foody_HaiDuong.json"
```

```python
df = pd.read_csv(input)
display(df)
```

```python
def split_workings_hour(working_hours):
    if working_hours in ["Null", "Cả ngày", "", None]:
        return {
            'open': None,
            'close': None
        }
    
    shifts = working_hours.split("|")
    times = [shift.strip().replace(":", "").split(' - ') for shift in shifts]
    
    # Nếu không có ca làm việc hợp lệ
    if not times or len(times[0]) < 2:
        return {
            'open': None,
            'close': None
        }

    # Xử lý giờ mở cửa và giờ đóng cửa theo số lượng ca làm việc
    if len(times) == 1:  # Chỉ có 1 ca làm việc
        open_time = times[0][0] if len(times[0]) > 0 else None
        close_time = times[0][1] if len(times[0]) > 1 else None
    else:  # Có nhiều ca làm việc, lấy giờ đầu và cuối
        open_time = times[0][0]
        close_time = times[-1][1] if len(times[-1]) > 1 else None

    return {
        'open': open_time,
        'close': close_time
    }
```

```python
time_data = df['Working hours'].apply(split_workings_hour)
df_time = pd.DataFrame(list(time_data))
df = pd.concat([df, df_time], axis=1)
# df['tags'] = df['City'].apply(lambda x: unidecode(x).lower().replace(' ', ''))
display(df)
```

```python
def row_to_json(row):
    return {
        "location": {
            "lng": row['Longitude'],
            "lat": row['Latitude']
        },
        "name": row['Name.1'],
        "objectId": None,
        "description": None,
        "types": ["food_service", "point"],
        "tags": ["hai_duong"],
        "address": row['Address'],
        "photos": None,
        "startDate": "2025-04-03T04:06:01.534Z",
        "endDate": "2025-04-03T04:06:01.534Z",
        "phoneNumber": None,
        "website": None,
        "businessHours" : [
            {
                "open" : {
                    "day" : 0, 
                    "time" : row['open']
                }, 
                "close" : {
                    "day" : 0, 
                    "time" : row['close']
                }
            }
        ], 
        "geometry": {
            "type": "Point",
            "coordinates": [row['Longitude'], row['Latitude']]
        },
        "rank": {
            "value": 0
        },
        "layer": "address",
        "source": {
            "name": "foody"
        },
        "metadata": None
    }

json_data = df.apply(row_to_json, axis=1).tolist()
print(json.dumps(json_data, indent=2, ensure_ascii=False))
```

```python
with open(output, "w", encoding="utf-8") as f:
    json.dump(json_data, f, indent=2, ensure_ascii=False)
```
