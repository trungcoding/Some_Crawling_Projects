```python
hotel_URL = []
list_name = []
list_address = []
list_latitude = []
list_longitude = []
```

```python
page_source = BeautifulSoup(driver.page_source, 'html.parser')
hotel_div = page_source.find_all('div', class_='PropertyCard__Link')
for hotel in hotel_div:
    href = hotel.get('data-element-url')  
    if href:
        hotel_URL.append(f"https://www.agoda.com/vi-vn{href}")
print(len(hotel_URL))
print(hotel_URL)
```

```python
hotel_URL = set(hotel_URL)
print(len(hotel_URL))
```

```python
# total_url = len(hotel_URL)
# for index, hotel in enumerate(hotel_URL):

start_index = 893
total_url = len(hotel_URL)
hotel_URL = list(hotel_URL)
for index, hotel in enumerate(hotel_URL[start_index:], start=start_index):
    try:
        driver.get(hotel)
        time.sleep(1)
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')

        name_tag = soup.find('meta', {'property': 'og:image:alt'})
        name = name_tag['content'] if name_tag and 'content' in name_tag.attrs else None

        # address_tag = soup.find('meta', {'property': 'og:street_address'})
        # address = address_tag['content'] if address_tag and 'content' in address_tag.attrs else None
        address_span = soup.find('span', class_='HeaderCerebrum__Address')
        address = address_span.get_text(strip=True)

        latitude_tag = soup.find('meta', {'property': 'place:location:latitude'})
        latitude = latitude_tag['content'] if latitude_tag and 'content' in latitude_tag.attrs else None

        longitude_tag = soup.find('meta', {'property': 'place:location:longitude'})
        longitude = longitude_tag['content'] if longitude_tag and 'content' in longitude_tag.attrs else None

        list_name.append(name)
        list_address.append(address)
        list_latitude.append(latitude)
        list_longitude.append(longitude)

        print(f"Index:{index+1}/{total_url} - {name} - {address} - {latitude} - {longitude}")

    except Exception as e:
        print(f"Li khi x l URL {hotel}: {e}")
        list_name.append("N/A")
        list_address.append("N/A")
        list_latitude.append("N/A")
        list_longitude.append("N/A")
```
