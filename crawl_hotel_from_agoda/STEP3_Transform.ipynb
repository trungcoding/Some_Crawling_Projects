```python
import pandas as pd
import csv
import json
```

```python
total_csv = "DATA2/ThemMoi_Hotel_TienGiang.csv"
update_json = "DATA2/ThemMoi_Hotel_TienGiang.json"
```

```python
total_df = pd.read_csv(total_csv)
display(total_df)
```

```python
def row_to_json(row):
    return {
        "location": {
            "lng": row['Longitude'],
            "lat": row['Latitude']
        },
        "name": row['Name.1'],
        "objectId": None,
        "description": None,
        "types": ["hotel", "lodging", "point"],
        "tags": ["tien_giang"],
        "address": row['Address'],
        "photos": None,
        "startDate": "2025-04-03T04:06:01.534Z",
        "endDate": "2025-04-03T04:06:01.534Z",
        "phoneNumber": None,
        "website": None,
        "businessHours" : None, 
        "geometry": {
            "type": "Point",
            "coordinates": [row['Longitude'], row['Latitude']]
        },
        "rank": {
            "value": 0
        },
        "layer": "address",
        "source": {
            "name": "agoda"
        },
        "metadata": None
    }

add = total_df.apply(row_to_json, axis=1).tolist()
print(json.dumps(add, indent=2, ensure_ascii=False))
```

```python
with open(update_json, "w", encoding="utf-8") as f:
    json.dump(add, f, indent=2, ensure_ascii=False)
```
