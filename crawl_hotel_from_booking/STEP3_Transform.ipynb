```python
import pandas as pd
import csv
import json
```

```python
total_csv = "D:/TRUNG/Booking/Booking_TraVinh.csv"
# update_csv = "SoSanh_Hotel_HoChiMinh.csv"
# add_json = "ADD_hotel_HoChiMinh.json"
update_json = "Booking_TraVinh.json"
```

```python
total_df = pd.read_csv(total_csv)
# update_df = pd.read_csv(update_csv)
display(total_df)
# display(len(update_df))
```

```python
add_df = total_df[
    ~total_df.set_index(['Latitude', 'Longitude']).index.isin(update_df.set_index(['Latitude', 'Longitude']).index)
].reset_index(drop=True)
display(add_df)
```

```python
def row_to_json(row):
    return {
        "location": {
            "lng": row['Longitude'],
            "lat": row['Latitude']
        },
        "name": row['Hotel Name'],
        "objectId": None,
        "description": None,
        "types": ["hotel", "lodging", "point"],
        "tags": ["tien_giang"],
        "address": row['Address'],
        "photos": None,
        "startDate": "2025-02-10T03:57:07.785Z",
        "endDate": "2025-02-10T03:57:07.785Z",
        "phoneNumber": None,
        "website": row['Link'],
        "businessHours" : None, 
        "geometry": {
            "type": "Point",
            "coordinates": [row['Longitude'], row['Latitude']]
        },
        "rank": {
            "value": 0
        },
        "layer": "address",
        "source": {
            "name": "booking"
        },
        "metadata": None
    }

add = total_df.apply(row_to_json, axis=1).tolist()
print(json.dumps(add, indent=2, ensure_ascii=False))
```

```python
with open(update_json, "w", encoding="utf-8") as f:
    json.dump(add, f, indent=2, ensure_ascii=False)
```
