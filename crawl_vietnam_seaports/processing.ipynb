```python
import pandas as pd 
import csv
import re
```

```python
df = pd.read_csv("D:/TRUNG/Cảng biển/VinaMarine.csv", header=1)
display(df)
```

```python
df.insert(df.columns.get_loc('GoogleMap'), 'Latitude', None)
df.insert(df.columns.get_loc('GoogleMap'), 'Longitude', None)
display(df)
```

```python
for index, row in df.iterrows():
    if pd.notna(row['GoogleMap']):  # Kiểm tra nếu không phải NULL
        match = re.search(r'!3d([0-9\.\-]+)!4d([0-9\.\-]+)', row['GoogleMap'])
        if match:
            df.at[index, 'Latitude'] = float(match.group(1))
            df.at[index, 'Longitude'] = float(match.group(2))
display(df)
df.to_csv("D:/TRUNG/Cảng biển/VinaMarine_Updated.csv", index=False)
```
